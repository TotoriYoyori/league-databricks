```sql
USE CATALOG league_records;

USE SCHEMA silver;

SELECT DISTINCT game_version
FROM bronze.matches
;

SELECT
    UPPER(match_id) AS match_id,
    TRY_CAST(ROUND(
        TRY_CAST(game_duration AS DOUBLE)
    , 0) AS INT) AS game_duration,
    CASE winning_team
        WHEN '100' THEN 'Blue'
        WHEN '200' THEN 'Red'
        ELSE NULL
    END AS winning_team,
    TRY_CAST(game_date AS TIMESTAMP) AS game_date,
    game_version,
    INITCAP(average_rank) AS average_rank,
    TRANSFORM(
        SPLIT(blue_bans, ','),
        x -> CASE WHEN TRY_CAST(x AS INT) = -1 THEN 0 ELSE TRY_CAST(x AS INT) END
    ) AS blue_bans,
    TRANSFORM(
        SPLIT(red_bans, ','),
        x -> CASE WHEN TRY_CAST(x AS INT) = -1 THEN 0 ELSE TRY_CAST(x AS INT) END
    ) AS red_bans
FROM bronze.matches
WHERE game_version IS NULL
;
```
